```python
import zipfile
from bs4 import BeautifulSoup

def inspect_docx_structure(docx_file):
    with zipfile.ZipFile(docx_file, 'r') as zip_ref:
        xml_content = zip_ref.read('word/document.xml')
        soup = BeautifulSoup(xml_content, 'xml')

        for tag in soup.find_all():
            print(tag.name, tag.attrs)

        # 特定の要素を検索 (例: 段落、表、画像)
        # paragraphs = soup.find_all('w:p')
        # tables = soup.find_all('w:tbl')
        # images = soup.find_all('w:drawing')

        # print("段落の数:", len(paragraphs))
        # print("表の数:", len(tables))
        # print("画像の数:", len(images))

        # フッターの確認
        footers = soup.find_all('w:footer')
        for footer in footers:
            print("フッター:", footer)
```

```python
# 使用例
inspect_docx_structure("/content/sample14.docx")

import zipfile
from bs4 import BeautifulSoup

def extract_and_save_elements(docx_file, output_file):
    with zipfile.ZipFile(docx_file, 'r') as zip_ref:
        xml_content = zip_ref.read('word/document.xml')
        soup = BeautifulSoup(xml_content, 'xml')

        with open(output_file, 'w', encoding='utf-8') as f:
            for tag in soup.find_all():
                # 書き出す情報のカスタマイズ
                # 例: タグ名と属性をタブ区切りで書き出す
                f.write(f"{tag.name}\t{tag.attrs}\n")

# 使用例
extract_and_save_elements("/content/sample14.docx", "extracted_elements.txt")
```

**↓のコードでテキストのみを抽出**

```python
import zipfile
from bs4 import BeautifulSoup

def extract_text(docx_file, output_file):
    with zipfile.ZipFile(docx_file, 'r') as zip_ref:
        xml_content = zip_ref.read('word/document.xml')
        soup = BeautifulSoup(xml_content, 'xml')

        with open(output_file, 'w', encoding='utf-8') as f:
            for paragraph in soup.find_all('w:p'):
                f.write(paragraph.text + '\n')

# 使用例
extract_text("/content/sample14.docx", "extracted_text.txt")
```
